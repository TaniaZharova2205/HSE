Программа должна вывести одно целое число.

```python
a=int(input())
b=int(input())
c=int(input())
i=0
count_iter = min(a, b//2, c)
i=count_iter*4
if a - count_iter>0:
    i+=1
    if b - count_iter*2>0:
        i+=1
        if c - count_iter>0:
            i+=1
print(i)
```

На уроке информатики Васе рассказали про сортировку натуральных чисел и он решил придумать метрику, по которой он мог бы судить, насколько последовательность отсортирована. Для этого он решил считать, сколько элементов последовательности больше, чем предыдущий элемент. Напишите программу, решающую эту задачу.

*Формат ввода*

Вводится последовательность целых чисел, оканчивающаяся нулем (сам 0 в последовательность не входит).

*Формат вывода*

Программа должна вывести сколько элементов введенной последовательности больше предыдущего элемента.

```python
pred_number = int(input())
count_ = 0
while pred_number!=0:
    now_number = int(input())
    count_ += 1 if now_number>pred_number else 0
    pred_number = now_number
print(count_)
```

В последовательности записаны целые числа от 1 до N в произвольном порядке, но одно из чисел пропущено (остальные встречаются ровно по одному разу). N заранее неизвестно. Определить пропущенное число

*Формат ввода*

На вход программе сначала подается натуральное n: $(1≤n≤10^5)$ – количество элементов в массиве.

Далее, в строке расположены сами элементы массива – целые числа, по модулю не превосходящие $3*10^4$

*Формат вывода*

Единственное число, которое пропущено в массиве.

```python
n = int(input())
lst = set(map(int, input().split()))
cnt = 0
for i in range(1, n+1):
    if i not in lst:
        cnt = i
        break
if cnt == 0:
    cnt = n+1
print(cnt)
```

Напишите программу, которая инвертирует определенный бит в заданном числе (биты при этом нумеруются с 0, начиная с младших).

*Формат ввода*

Дано целое число A и натуральное число i.

*Формат вывода*

Выведите число, которое получается из числа A инвертированием i-го бита.

```python
a, i = map(int, input().split())
a_2 = bin(a)[2:]
len_a = len(a_2)
if len_a <= i:
    a_2 = '0' * (i - len_a + 1) + a_2
a_2 = a_2[::-1]
a_2 = a_2[:i] + '0' + a_2[i+1:] if a_2[i] == '1' else a_2[:i] + '1' + a_2[i+1:]
print(int(a_2[::-1], 2))
```

Вася придумал новую метрику по которой он может делать выводы о том, насколько отсортирована последовательность натуральных чисел. Для этого он находит длину наибольшей монотонной подпоследовательности. Монотонной подпоследовательностью считается такой фрагмент последовательности, в котором все элементы располагаются либо в порядке возрастания, либо в порядке убывания. Напишите программу, вычисляющую эту метрику.

*Формат ввода*

Вводится последовательность целых чисел, оканчивающаяся нулем (сам 0 в последовательность не входит).

*Формат вывода*

Программа должна вывести длину максимальной монотонной подпоследовательности.

```python
pred_number = int(input())
count_down = 1
count_up = 1
max_count = 1
while True:
    now_number = int(input())
    if now_number == 0:
        break
    if now_number>pred_number:
        count_down += 1
        count_up = 1
        max_count = max(count_down, max_count)
    elif now_number<pred_number:
        count_up += 1
        count_down = 1
        max_count = max(max_count, count_up)
    else:
        count_down = 1
        count_up = 1
    pred_number = now_number
print(max_count)
```
