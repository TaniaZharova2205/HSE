```python
import time 
import matplotlib.pyplot as plt

#функция умножения матриц list
def dot_AB(n, A, B): 
    A_dot_B = [[0 for i in range(n)] for i in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                A_dot_B[i][j] += A[i][k]*B[k][j]
    return A_dot_B

#время выполнения numpy
def measure_numpy_multiplication(A,B):
    time_start = time.time()
    A.dot(B)
    return time.time() - time_start

#время выполнения list
def measure_list_multiplication(n,A,B):
    time_start = time.time()
    dot_AB(n, A, B)
    return time.time() - time_start

#отрисовка графиков
def graph(x, time_numpy, time_list):
    
    fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(25, 6))
    
    axes[0].plot(x, time_numpy, color="#00FFFF", label = "numpy")
    axes[0].plot(x, time_list, color="#FF69B4", label = "списки")
    axes[0].set_xlabel("Размер матрицы")
    axes[0].set_ylabel("Время выполнения умножения матриц")
    axes[0].legend()
    
    axes[1].plot(x, time_numpy, color="#00FFFF", label = "numpy")
    axes[1].set_xlabel("Размер матрицы")
    axes[1].set_ylabel("Время выполнения умножения матриц с помощью numpy")
    
    axes[2].plot(x, time_list, color="#FF69B4", label = "списки")
    axes[2].set_xlabel("Размер матрицы")
    axes[2].set_ylabel("Время выполнения умножения матриц с помощью списков")

n = np.arange(2,101)
time_numpy = []
time_list = []
for i in n:
    A = np.random.randint(1, 51, (i, i))
    B = np.random.randint(1, 51, (i, i))

    A_list = A.tolist()
    B_list = B.tolist()
    
    time_numpy.append(measure_numpy_multiplication(A,B))
    time_list.append(measure_list_multiplication(i, A_list,B_list))

    #проверка на совпадение результатов
    if i == 88:
        print(np.array_equal(A.dot(B), np.array(dot_AB(i, A, B))))

graph(n, time_numpy, time_list)
```

```python
def salary_num_function(value):
    if value == {}:
        return 0
    elif value['from']!='null' and value['to']!='null':
        return (value['to'] - value['from'])//2
    elif value['from']!='null':
        return value['from']
    elif value['to']!='null':
        return value['to']

df['salary_num'] = df['salary'].apply(salary_num_function)
df['salary_num'].head()
```

```python
def is_python_requirement_function(x):
    for i in x:
        if i.lower() == 'python':
            return 1
    return 0
df['is_python_requirement'] = df['key_skills'].apply(is_python_requirement_function)
print(df[df['is_python_requirement']==1]['is_python_requirement'].count()*100/df.shape[0])
df.head()
```

```python
def salary_num_RUR(x):
    if x['currency'] == 'RUR':
        return x
    else:
        x['salary_num'] = 0
        return x

rur_salaries = df.apply(salary_num_RUR, axis=1)
rur_salaries.groupby(['experience', 'schedule', 'employment', 'is_python_requirement'], as_index=False).agg(
    count_vacancy=('experience', 'count'),
    salary_mean = ('salary_num', 'mean'),
    count_employer_name = ('employer_name', 'nunique')
)
```

```python
table_pivot = pd.pivot_table(rur_salaries,
               index = 'employer_name',
               values=['position_name', 'salary_num'],
               aggfunc={
                    'position_name': 'count',
                    'salary_num': 'mean'
                },
            )
table_pivot.rename(columns={'position_name': 'count_vacancy', 'salary_num': 'salary_num_mean'})
```

```python
def scatter_plot(df, col1, col2, title, xlab, ylab):
    plt.figure(figsize=(5, 5))
    plt.title(title)
    plt.scatter(df[col1], df[col2])
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.show()
numeric_cols = df.select_dtypes(include='number').columns
for i in range(len(numeric_cols)):
    for j in range(i+1, len(numeric_cols)):
        scatter_plot(df, numeric_cols[i], numeric_cols[j], f"{numeric_cols[i]} vs {numeric_cols[j]}", numeric_cols[i], numeric_cols[j])
```
